```python
# Lab program 1:
#Write a Program to read a digital image. 
#Split and display image into 4 quadrants, up, down, right &left
import cv2 as cv
import numpy as np
image=cv.imread('/home/user/Pictures/daya_DIP_lab/chery-200x141.png')
#Converts from BGR to RGB

image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

(h, w) = image.shape[:2]

# compute the center coordinate of the image
(cX, cY) = (w // 2, h // 2)
# crop the image into four parts which will be labelled as
# top left, top right, bottom left, and bottom right.
topLeft = image[0:cY, 0:cX]
topRight = image[0:cY, cX:w]
bottomLeft = image[cY:h, 0:cX]
bottomRight = image[cY:h, cX:w]
#import libraries

import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt

# create figure
fig1 = plt.figure(figsize=(4,4))
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
fig = plt.figure(figsize=(4,4))

# setting values to rows and column variables
rows = 2
columns = 2


# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)

# showing image
plt.imshow(topLeft)
plt.axis('off')
plt.title("Top left part")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

# showing image
plt.imshow(topRight)
plt.axis('off')
plt.title("Top right part")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)

# showing image
plt.imshow(bottomLeft)
plt.axis('off')
plt.title("Bottom left part")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)

# showing image
plt.imshow(bottomRight)
plt.axis('off')
plt.title("Bottom right part")
plt.show()
```
